# Broadcasting en NumPy

Tres ejemplos reales de broadcasting pensados para alumnos básicos:

1. Broadcasting de dos vectores para crear una matriz
2. Broadcasting en un eje
3. Broadcasting en dos ejes

## Ejemplo 1 — Dos vectores → matriz

Coste total según precios y cantidades.

```python
import numpy as np

precios = np.array([2.5, 4.0, 6.0])
cantidades = np.array([10, 20, 15, 30, 25])

coste_total = precios[:, np.newaxis] * cantidades

coste_total
```

## Ejemplo 2 — Broadcasting en un eje

Ventas de varias tiendas con impuesto por tienda.

```python
ventas = np.array([
    [100, 120, 90, 110, 130],
    [80, 95, 85, 100, 105],
    [200, 210, 190, 205, 215]
])

impuestos = np.array([1.10, 1.08, 1.21])

ventas_con_impuesto = ventas * impuestos[:, np.newaxis]

ventas_con_impuesto
```

## Ejemplo 3 — Broadcasting en dos ejes

Modificación de una imagen aplicando factores por fila y columna.

```python
imagen = np.array([
    [50, 60, 70, 80],
    [90, 100, 110, 120],
    [130, 140, 150, 160],
    [170, 180, 190, 200]
])

brillo_fila = np.array([0.8, 1.0, 1.1, 1.3])
contraste_columna = np.array([1.0, 0.9, 1.1, 1.2])

imagen_modificada = imagen * brillo_fila[:, np.newaxis] * contraste_columna

imagen_modificada
```
