```python
def sumar(*args):
    print("args recibido:", args)
    return sum(args)

resultado = sumar(1, 2, 3, 4)
print("Resultado:", resultado)
```

```python
def mostrar_info(**kwargs):
    print("kwargs recibido:", kwargs)
    for clave, valor in kwargs.items():
        print(f"{clave} = {valor}")

mostrar_info(nombre="Ana", edad=28, ciudad="Madrid")
```

```python
def configurar_servidor(host="localhost", port=8000, **extras):
    print(f"Servidor en {host}:{port}")
    print("Extras:", extras)

configurar_servidor(debug=True, workers=4, reload=True)
```

```python
def query(sql, **params):
    print("Ejecutando:", sql, "con", params)

query("SELECT * FROM users WHERE id=:id", id=10)
```
