# Function as argument

```python
def do_twice(func):
    func()
    func()

def print_one():
    print("one")

do_twice(print_one)
```

# Inner functions

```python
def parent():
    print("Printing from parent()")

    def first_child():
        print("Printing from first_child()")

    def second_child():
        print("Printing from second_child()")

    second_child()
    first_child()

parent()
```

```python
first_child()
```

# Returning functions

```python
def parent(num):
    def first_child():
        return "Hi, I'm Elias"

    def second_child():
        return "Call me Ester"

    if num == 1:
        return first_child
    else:
        return second_child
```

```python
first = parent(1)
second = parent(2)

second
```

```python
first()
```

# Decorators

```python
def decorador(func):
    def wrapper():
        print("Antes.")
        func()
        print("Despues.")
    return wrapper

def print_hola():
    print("Hola")

hola_decorado = decorador(print_hola)

hola_decorado()
```

```python
@decorador
def adios():
    print("adios")

adios()
```

## Ejemplo de decorador en el mundo real

```python
import functools
import time

def slow_down(*, seconds=1):
    def decorator(func):
        @functools.wraps(func)
        def wrapper(*args, **kwargs):
            time.sleep(seconds)
            return func(*args, **kwargs)
        return wrapper
    return decorator

@slow_down(seconds=5)
def print_inputs(*args,**kwargs):
    print(f"args: {args}")
    print(f"kwargs: {kwargs}")

print_inputs("hola","que","tal", hola="hola", adios="adios")
```

Esto es equivalente a:

```python
def print_inputs(*args,**kwargs):
    print(f"args: {args}")
    print(f"kwargs: {kwargs}")


print_inputs = slow_down(seconds=5)(print_inputs)

print_inputs("hola")
```
